```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import os

path = "C:\\Users\\Kevin_Andrew\\Desktop\\Kevin_Andrew_Research\\MinWage\\Data_Work\\HIPRIP"

os.chdir(path)

empdata = pd.read_excel('empcov.xlsx', index_col = None, header = None)
popdatarip = pd.read_excel('popcovrip.xlsx', index_col = None, header = None)
popdatahip = pd.read_excel('popcovhip.xlsx', index_col = None, header = None)

covmat    = empdata.values
popmatrip = popdatarip.values
popmathip = popdatahip.values

amax = 15

for ia in range(amax):
    agrid = np.linspace(ia+1, amax+1, amax - ia)
    aplot = covmat[ia,ia:amax]
    pplot = popmatrip[ia,ia:amax]
    plt.plot(agrid, aplot, 'b-', linewidth = 3)
    plt.plot(agrid, pplot, 'g-', linewidth = 3)
plt.title('Performance of the RIP Model')
plt.show()

for ia in range(amax):
    agrid = np.linspace(ia+1, amax+1, amax - ia)
    aplot = covmat[ia,ia:amax]
    pplot = popmathip[ia,ia:amax]
    plt.plot(agrid, aplot, 'b-', linewidth = 3)
    plt.plot(agrid, pplot, 'g-', linewidth = 3)
plt.title('Performance of the HIP Model')
plt.show()
   
```
